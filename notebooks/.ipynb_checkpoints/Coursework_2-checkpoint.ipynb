```python
import numpy as np
from collections import OrderedDict
import logging
from mlp.layers import SoftmaxLayer, AffineLayer, ReluLayer, SigmoidLayer, TanhLayer, BatchNormLayer, ConvolutionalLayer, MaxPoolingLayer, ReshapeLayer, DropoutLayer
from mlp.penalties import L1Penalty, L2Penalty
from mlp.errors import CrossEntropySoftmaxError
from mlp.models import MultipleLayerModel
from mlp.initialisers import GlorotUniformInit, ConstantInit, UniformInit
from mlp.learning_rules import MomentumLearningRule, GradientDescentLearningRule
from mlp.data_providers import MNISTDataProvider
from mlp.optimisers import Optimiser

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

# Seed a random number generator
seed = 24102016 
rng = np.random.RandomState(seed)

num_epochs = 100
stats_interval = 5
batch_size = 50
learning_rate = 0.05
mom_coeff = 0.9
weights_init_gain = 0.5
kernels_init_gain = 0.5
biases_init = 0.

# Set up a logger object to print info about the training run to stdout
logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.handlers = [logging.StreamHandler()]

# Create data provider objects for the MNIST data set
train_data = MNISTDataProvider('train', batch_size, rng=rng)
valid_data = MNISTDataProvider('valid', batch_size, rng=rng)
input_dim, output_dim, hidden_dim = 784, 10, 100

weights_init = GlorotUniformInit(weights_init_gain, rng)
kernels_init = UniformInit(-kernels_init_gain, kernels_init_gain)
biases_init = ConstantInit(biases_init)
error = CrossEntropySoftmaxError()
learning_rule = MomentumLearningRule(learning_rate, mom_coeff)
data_monitors = {'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
```

```python
def make_plot2(run1, log=False, fig_name=None):
    fig = plt.figure(figsize=(12, 6))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    for transfer_function, run in run1.items():
        stats, keys, run_time = run
        ax1.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                 stats[1:, keys['error(train)']], label=str(transfer_function))
        ax2.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                 stats[1:, keys['error(valid)']], label=str(transfer_function))
    ax1.legend(loc=0)
    ax1.set_ylim([0.0,0.8])
    ax2.set_ylim([0.0,0.8])
    ax1.set_xlim([0,100])
    ax2.set_xlim([0,100])
    ax1.set_xlabel('Epoch number')
    ax1.set_ylabel('Training set error')
    if log:
        ax1.set_yscale('log')
    ax2.legend(loc=0)
    ax2.set_xlabel('Epoch number')
    ax2.set_ylabel('Validation set error')
    if log:
        ax2.set_yscale('log')
    if fig_name is not None:
        fig.tight_layout()
        fig.savefig(fig_name)
```

```python
def make_plot_acc(run1, log=False, fig_name=None):
    fig = plt.figure(figsize=(12, 6))
    ax1 = fig.add_subplot(1, 2, 1)
    ax2 = fig.add_subplot(1, 2, 2)
    ax1.set_ylim([0.9,1.02])
    ax2.set_ylim([0.9,1.02])
    for transfer_function, run in run1.items():
        stats, keys, run_time = run
        ax1.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                 stats[1:, keys['acc(train)']], label=str(transfer_function))
        ax2.plot(np.arange(1, stats.shape[0]) * stats_interval, 
                 stats[1:, keys['acc(valid)']], label=str(transfer_function))
    ax1.legend(loc=0)
    ax1.set_xlabel('Epoch number')
    ax1.set_ylabel('Training set acc')
    if log:
        ax1.set_yscale('log')
    ax2.legend(loc=0)
    ax2.set_xlabel('Epoch number')
    ax2.set_ylabel('Validation set acc')
    if log:
        ax2.set_yscale('log')
    if fig_name is not None:
        fig.tight_layout()
        fig.savefig(fig_name)
```

```python
def train_with_transfer_functions(transfer_functions):
    
    run_info = OrderedDict()
    weights_init = GlorotUniformInit(rng=rng)
    
    for transfer_function in transfer_functions:
        
        # Reset random number generator and data provider states on each run
        # to ensure reproducibility of results
        rng.seed(seed)
        train_data.reset()
        valid_data.reset()
        
        print('Transfer Function: {}'.format(transfer_function))
        
        model = MultipleLayerModel([
            AffineLayer(input_dim, hidden_dim, weights_init, 
                        biases_init, L2Penalty(1e-4)),
            transfer_function,
            AffineLayer(hidden_dim, hidden_dim, weights_init, 
                        biases_init, L2Penalty(1e-4)),
            transfer_function,
            AffineLayer(hidden_dim, output_dim, weights_init, 
                        biases_init, L2Penalty(1e-4)),
        ])
        error = CrossEntropySoftmaxError()
        
        learning_rule = MomentumLearningRule(learning_rate=0.01, mom_coeff=0.9)
        data_monitors = {'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
        
        optimiser = Optimiser(model, error, learning_rule, train_data, valid_data, data_monitors)
        run_info[str(transfer_function)] = optimiser.train(num_epochs, stats_interval)
    return run_info
```

```python
transfer_functions = [
    ReluLayer(),
    TanhLayer(),
    SigmoidLayer()
]
run55 = train_with_transfer_functions(transfer_functions)
```

```python
def train_model(scales):
    
    run_info = OrderedDict()
    
    for scale in scales:
    
        # Reset random number generator and data provider states on each run
        # to ensure reproducibility of results
        rng.seed(seed)
        train_data.reset()
        valid_data.reset()
    
        print('Filter Scale :{}'.format(scale))
        kernels_init = UniformInit(-scale, scale)
        
        model = MultipleLayerModel([
            ReshapeLayer((1, 28, 28)),
            ConvolutionalLayer(
                num_input_channels=1,
                num_output_channels=3,
                input_dim_1=28, input_dim_2=28,
                kernel_dim_1=5, kernel_dim_2=5,
                kernels_init=kernels_init
            ),
            ReshapeLayer((-1,)),
            ReluLayer(),
            MaxPoolingLayer(4),
            ReshapeLayer((3,12,12)),    
            ConvolutionalLayer(
                num_input_channels=3,
                num_output_channels=6,
                input_dim_1=12, input_dim_2=12,
                kernel_dim_1=5, kernel_dim_2=5,
                kernels_init=kernels_init
            ),
            ReshapeLayer((-1,)),
            ReluLayer(),
            MaxPoolingLayer(2),
            AffineLayer(192, hidden_dim, weights_init, biases_init),
            ReluLayer(),
            AffineLayer(hidden_dim, output_dim, weights_init, biases_init)
        ])
        
        error = CrossEntropySoftmaxError()

        learning_rule = MomentumLearningRule(learning_rate=0.001, mom_coeff=0.9)
        data_monitors = {'acc': lambda y, t: (y.argmax(-1) == t.argmax(-1)).mean()}
        
        optimiser = Optimiser(model, error, learning_rule, train_data, valid_data, data_monitors)
        run_info['Scale {}'.format(scale)] = optimiser.train(40, stats_interval)
    
    return run_info
```

```python
_scales3 = [1.0]
conv_res3 = train_model(_scales3)
```

```python
_scales = [0.001, 0.01, 0.05, 0.5]
conv_res = train_model(_scales)
```

```python
def train_with_batch(layers, no_bn=False):
    
    run_info = OrderedDict()
    weights_init = GlorotUniformInit(rng=rng)
    gamma = 1.0
    beta = 0.0
    
    for num_layers in layers:
        
        # Reset random number generator and data provider states on each run
        # to ensure reproducibility of results
        rng.seed(seed)
        train_data.reset()
        valid_data.reset()
            
        model_layers = [
            AffineLayer(input_dim, hidden_dim, weights_init, biases_init)
        ]
        
        for i in range(num_layers):
            #model_layers.append(BatchNormLayer(gamma, beta, hidden_dim))
            model_layers.append(ReluLayer())   
            model_layers.append(AffineLayer(hidden_dim, hidden_dim, weights_init, biases_init))
        
        #model_layers.append(BatchNormLayer(gamma, beta, hidden_dim))    
        model_layers.append(ReluLayer())
        model_layers.append(AffineLayer(hidden_dim, output_dim, weights_init, biases_init))
        
        print('{} Layers'.format(num_layers+1))
        model = MultipleLayerModel(model_layers)
        error = CrossEntropySoftmaxError()
        
        learning_rule = MomentumLearningRule(0.001)
            
        optimiser = Optimiser(model, error, learning_rule, train_data, valid_data, data_monitors)
        run_info['{} Layers'.format(num_layers+1)] = optimiser.train(100, stats_interval)
    return run_info
```
